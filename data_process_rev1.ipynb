# About this script

This can enable you to make graph using csv files in a certain folder.
And csv should have 2 cols. if all units in csv file were same, you could get one graph including every data.

## necessary library

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
import os
%matplotlib inline
```

## folder selection

```python
# to get csv file here and put them into a list 'list_data'
folder = os.getcwd()
target_extension = '*.csv'
mypath = Path(folder + '/data')
list_data = list(mypath.glob(target_extension))
```

## collect data from csv files

- title of dataframe
- array of dataframe
- columns information of dataframe

```python
file_num = len(list_data)

df_array = [] # dataframe using csv file
df_list = [] # column name of dataframe
df_file_title = [] # csv file name of dataframe
for x in list_data:
    df_file_title.append(x.name)
    tmp = pd.read_csv(x, header=0)
    df_array.append(tmp)
    df_list.append(tmp.columns)
```

## graph generation

```python
fig, axes = plt.subplots(1, 1, tight_layout=True, squeeze=False, edgecolor='white', linewidth=1)

# read dataframe and put them into one axes object.
for x in range(file_num):
    df_array[x].plot(x=df_list[x][0], y=df_list[x][1], ax=axes[0,0], label=df_file_title[x])

axes[0,0].set_title('title')
axes[0,0].set_xlabel('x axis')
axes[0,0].set_ylabel('y axis')
axes[0,0].set_facecolor('whitesmoke')
axes[0,0].tick_params(direction='out', axis='both', width=2)
axes[0,0].grid(visible=None, which='major', axis='both', linewidth=1, linestyle='-.')

plt.show()
```
